```python
# stemming
words = ["run", "runner", "running", "ran", "runs", "eats", "eating", "ate", "eat"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()

for word in words:
    print(f"{word} -> {stemming.stem(word)}")
```

```python
# dissadvantage of stemming is that it can produce non-words, for example:
stemming.stem("happiness")
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer("ing$|s$|e$|able$|ible$", min=4)

reg_stemmer.stem("eating")
```

```python
for word in words:
    print(f"{word} -> {reg_stemmer.stem(word)}")
```

```python
# Snowball Stemmer
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer("english")
```

```python
for word in words:
    print(f"{word} -> {snowball_stemmer.stem(word)}")
```
